# Visualizar el mapa de Barcelona

Veremos como hacer para ver en pantalla el mapa de la ciudad de Barcelona que nos sirva para poder dibujar sobre él datos de interés para nuestro proyecto. Para poder hacerlo tendremos que utilizar algunas librerías auxiliares de Python.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import fiona
%matplotlib inline
```

Para dibujar mapas Python tiene una biblioteca llamada `basemap` que forma parte de un módulo auxiliar de la biblioteca `matplotlib`. Basemap permite leer mapas y dibujar sobre ellos símbolos o formas que corresponderán a los datos que nosotros queremos graficar.

El mapa que usaremos está en un formato típico de los mapas digitales conocido como "shapefile", que en realidad es un conjunto de archivos a través de los cuales se distribuye la información, algunos archivos contienen información geográfica y otros atributos varios.

Lo primero que haremos para poder hacer el gráfico de nuestro mapa es encontrar los límites de nuestro mapa de Barcelona. Para encontrar las coordenadas de esos límites usaremos el módulo `Fiona` como se muestra abajo:  

```python
# Leemos el archivo que contiene los datos geográficos del mapa
shp = fiona.open('data/mapas/area-estadistica/BCN_Area_Estadistica_Basica.shp')
# Extraemos los límites de nuestro mapa y los guardamos en la variable limites
limites = shp.bounds
# cerramos el archivo
shp.close()
# Las coordenadas de la esquina inferior izquierda es (x1,y1) 
# y las coordenadas de la esquina superior derecha es (x2,y2)
x1 = limites[0]
y1 = limites[1]
x2 = limites[2]
y2 = limites[3]
# Imprimimos los valores de los límites del mapa (latitudes y longitudes en grados)
print x1, y1, x2, y2
```

Ahora podemos crear nuestro mapa.

## Mapa con división por áreas estadísticas básicas

```python
plt.figure(figsize=(10,10))
m = Basemap(
    projection='tmerc',
    lon_0= (x1 + x2)/2,
    lat_0= (y1 + y2)/2,
    llcrnrlon = x1,
    llcrnrlat = y1,
    urcrnrlon = x2,
    urcrnrlat = y2,
    lat_ts=0,
    resolution='i',
    suppress_ticks=True)
m.readshapefile(
    'data/mapas/area-estadistica/BCN_Area_Estadistica_Basica',
    'BCN_Area_Estadistica_Basica',
    color='green',
    zorder=2)
```

Este mapa corresponde a las áreas estadísticas básicas.
