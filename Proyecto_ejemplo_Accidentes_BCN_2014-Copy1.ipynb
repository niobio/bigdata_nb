### Distribución de los accidentes en al mapa de Barcelona

Podemos utilizar la geolocalización de los accidentes para tratar de verlos sobre un mapa de la ciudad y que nos permita ver a simple vista si hay zonas con mayor concentración de accidentes.

Para trazar el mapa vamos a importar las bibliotecas necesarias:

```python
# modulo para hacer graficos
import matplotlib.pyplot as plt  
# modulo para que los mapas se puedan incluir en los graficos
from mpl_toolkits.basemap import Basemap
# modulo para extraer coordenadas de los mapas
import fiona 
# modulo que permite hacer transformaciones de coordenadas entre diferentes proyecciones
import mpl_toolkits.basemap.pyproj as pyproj
# hace que los graficos se dibujen dentro del notebook y no como ventana aparte
%matplotlib inline                          
```

El mapa base que utilizaremos será el que tiene la delimitación por distritos.

Las coordenadas geográficas de los registros están en coordenadas UTM, para poder usarlos tendremos que transformarlos primeros a coordenadas de latitud y longitud en grados. Eso se puede hacer con el módulo `pyproj` de la siguiente manera:

```python
# Primero extraemos los datos geográficos de las columnas
x = df['Coordenada UTM (X)']
y = df['Coordenada UTM (Y)']

# Aquí definimos las proyecciones 
# EPSG:23031 corresponde a UTM en la región donde se encuantra Barcelona
# EPSG:4326 corresponde a latitudes y longitudes
# Para +info: https://es.wikipedia.org/wiki/European_Petroleum_Survey_Group
p1=pyproj.Proj(init="epsg:23031")
p2=pyproj.Proj(init="epsg:4326")

# Aquí realizamos las transformaciones
# Observar que los datos contenidos en x e y son del tipo string 
# por los que hay que pasarlos a float pero antes hay que reemplazar 
# la coma decimal por un punto
longs = []
lats = []
for i in range(len(x)):
    coord = pyproj.transform(p1,p2,float(x[i].replace(',','.')),float(y[i].replace(',','.')))
    longs.append(coord[0])
    lats.append(coord[1])

# Verificamos que los datos corresponden a latitudes y longitudes de la
# zona de Barcelona imprimiendo los primeros elementos de la  lista
print longs[0], lats[0]
```

```python
# Leemos el archivo que contiene los datos geográficos del mapa
shp = fiona.open('data/mapas/districte/BCN_Districte.shp')
# Extraemos los límites de nuestro mapa y los guardamos en la variable limites
limites = shp.bounds
# cerramos el archivo
shp.close()

x1 = limites[0]
y1 = limites[1]
x2 = limites[2]
y2 = limites[3]

print x1,y1
print x2,y2

plt.figure(figsize=(12,12))
mapa = Basemap(
    projection='tmerc',
    lon_0= (x1 + x2)/2,
    lat_0= (y1 + y2)/2,
    llcrnrlon = x1,
    llcrnrlat = y1,
    urcrnrlon = x2,
    urcrnrlat = y2,
    lat_ts=0,
    resolution='i',
    suppress_ticks=True)
mapa.readshapefile(
    'data/mapas/districte/BCN_Districte',
    'BCN_Districte',
    color='green',
    zorder=2)

# Para poder poner los datos en el mapa se deben transformar 
# de latitud y longitud a coordenadas del mapa, para ello usamos
# el mismo objeto mapa
xm,ym = mapa(longs, lats)
mapa.plot(xm,ym,'ro', markersize= 2)
```

En el mapa podemos observar los datos globales de accidentes gestionados por la Guardia Urbana durante el año 2014. Vemos como las vías con mayor tránsito y donde el tránsito circula a mayor velocidad es donde se producen más accidentes. Para ver con mayor detalle se podría graficar el mapa eligiendo otros límites que seleccionen una parte más pequeña del mismo y de esa manera se podrían descubrir puntos negros de siniestralidad en la ciudad.

### ¿Qué día de la semana se producen más accidentes?

Para responder a esa pregunta lo mejor sería hacer un gráfico de barras en el que como coordenada horizontal (abscisas) tengamos los días de la semana y en el eje vertical (ordenadas) el número de accidentes correspondiente.

Lo primero que haremos será agrupar los datos por día de semana usando el método `groupby` de la clase `DataFrame`.

```python
gds  = df.groupby('Dia de setmana')
gds.ngroups
```

Vemos que los datos quedan agrupados en 7 grupos correspondientes a los días de la semana. Podemos saber cuantos elementos tiene cada grupo, y eso nos dirá el número de accidentes de cada día de la semana.

```python
gp = gds.size()
gp
```

El elemento gp no es un `DataFrame` sino un objeto `DataFrameGroupBy`, que tiene muhas utilidades. Para más información sobre cómo agrupar datos ver: http://pandas.pydata.org/pandas-docs/stable/groupby.html

Ahora trazamos el gráfico de barras:

```python
gp.plot(kind='bar')
```

Podemos también hacer también un gráfico de sectores que nos informe del porcentaje correspondiente a cada día de la semana:

```python
gp.plot(kind='pie')
''', 
        autopct='%.0f',
        labels=['Lunes','Martes',u'Miércoles','Jueves','Viernes',u'Sábado','Domingo'], 
        figsize=(6,6), 
        title='Accidentes 2014',
        label='Porcentaje',
        colors=['g','b','y','c','m','orange','r'],
        fontsize=12)'''
```

Vemos que los accidentes se reparten bastante equitativamente durante los días laborales y bajan durante el fin de semana debido a que hay menos vehículos circulando por la ciudad. Resulta curioso que los jueves y viernes aumenta levemente el número de accidentes respecto a los tres primeros días de la semana y podríamos hacernos la pregunta: ¿este fenómeno sucede para todos los años o es una característica puramente estadística del año 2014?
